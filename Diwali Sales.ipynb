# EDA And Data Visualisation

```python
Gender_amount=diwalisales_df["Gender"].value_counts()
Gender_percent=diwalisales_df["Gender"].value_counts(normalize=True)
print("Gender-Wise Amount")
print(Gender_amount)
print("="*100)
print("Gender-Wise Percent")
print(Gender_percent)

fig,ax=plt.subplots(1,1,figsize=(10,5))
ax=sns.countplot(x="Gender", data=diwalisales_df, ax=ax)
plt.title("Gender-Wise Count")
plt.ylabel("Amount of Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Male_Gender=diwalisales_df[diwalisales_df["Gender"]=="M"]
Female_Gender=diwalisales_df[diwalisales_df["Gender"]=="F"]
```

```python
Gender_Wise_Amount=diwalisales_df.groupby(['Gender'])['Amount'].sum()
print(Gender_Wise_Amount)
print("="*125)
fig,ax=plt.subplots(1,1,figsize=(10,5))
diwalisales_df.groupby(['Gender'])['Amount'].sum().plot(kind='bar')
plt.title("Total Price of Product Purchased")
plt.ylabel("Amount of Price")
```

```python
Gender_Wise_Order=diwalisales_df.groupby(['Gender'])['Orders'].sum()
print(Gender_Wise_Order)
print("="*125)

fig,ax=plt.subplots(1,1,figsize=(10,5))
diwalisales_df.groupby(['Gender'])['Orders'].sum().plot(kind='bar')
plt.title("Total Amount of Order Purchased")
plt.ylabel("Amount of Order")
```

```python
Male_Gender_Age_Per=Male_Gender['Age Group'].value_counts(normalize=True)
print(Male_Gender_Age_Per)
print("="*125)

fig,ax=plt.subplots(1,1,figsize=(10,5))
ax=sns.countplot(x="Age Group", data=Male_Gender,  ax=ax)
plt.ylabel("Number of Male Gender Per Age Group")
plt.title("Age_Group Wise Male Gender")

for bars in ax.containers:
    ax.bar_label(bars) 
```

```python
Female_Gender_Age_Per=Female_Gender['Age Group'].value_counts(normalize=True)
print(Female_Gender_Age_Per)
print("="*125)

fig,ax=plt.subplots(1,1,figsize=(10,5))
ax=sns.countplot(x="Age Group", data=Female_Gender,  ax=ax)
plt.ylabel("Number of female Gender Per Age Group")
plt.title("Age_Group Wise Female Gender")

for bars in ax.containers:
    ax.bar_label(bars) 
```

```python
Age_Group_Amount=diwalisales_df.groupby(['Age Group'])['Amount'].sum()
print(Age_Group_Amount)
print("="*125)

fig,ax=plt.subplots(1,1,figsize=(10,5))
ax=diwalisales_df.groupby(['Age Group'])['Amount'].sum().plot(kind='bar')
plt.title("Total Amount Spent Age_Group-Wise")
plt.ylabel("Amount Spent")

#sales_age=diwalisales_df.groupby(['Age Group'])['Amount'].sum().sort_values(key='Amount', ascending=False)
#sns.barplot(x="Age Group", y="Amount", data=sales_age)

for bars in ax.containers:
    ax.bar_label(bars) 
```
